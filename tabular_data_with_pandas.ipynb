### Data manipulation with Pandas 
* Select individual values from a Pandas dataframe.
* Select entire rows or entire columns from a dataframe.
* Select a subset of both rows and columns from a dataframe in a single operation.
* Select a subset of a dataframe by a single Boolean criterion.

#### Select individual value

```python
data = pandas.read_csv('data/gapminder_gdp_europe.csv', index_col='country')
#print(data.head())
#print(data.iloc[0,0])
print(data.iloc[0,3])

```

#### Select a row by index

```python
data.iloc[0,]
#data.iloc[0]
```

#### Select a column

```python
data.iloc[:,0]
#data.iloc[0]
```

#### Select n-rows (here n = 5 rows)

```python
print(data.iloc[0:5])
```

```python
#### Select 5 rows and 3 columns
```

```python
print(data.iloc[0:5,0:3])
```

#### Select an individual element by row and column names

```python
# specify location by row name 
print(data.loc['Albania', 'gdpPercap_1952'])
```

#### Select a row by name

```python
data.loc["Albania", :]
#data.loc["Albania"]
```

#### Select a column by name

```python
# first way
print(data.loc[:, "gdpPercap_1952"])
# second way
data["gdpPercap_1952"]
```

```python
#### Select rows and columns by slicing the names
```

```python
data.loc['Italy':'Poland', 'gdpPercap_1962':'gdpPercap_1972']
```

#### get maximum values in the dataframe subset

```python
data.loc['Italy':'Poland', 'gdpPercap_1962':'gdpPercap_1972'].max()
data.loc['Italy':'Poland', 'gdpPercap_1962':'gdpPercap_1972'].max(axis=1)
```
